## Problemas
1. El sistema usa RSA como mecanismo de intercambio de clave, protegiendo una clave AES que
cifra el documento real.
    
    a. ¿Explique por qué no cifrar el documento directamente con RSA?
    
    Porque RSA tiene limitaciónes en el cual solo cifra datos pequeños. Por ejemplo: 
    * RSA-2048 --> MÓDULO DE 256 BYTES 
    * con padding OAEP --> máximo 190 bytes de datos reales . 
    Por lo cual no sirve para cifrar archivos, imágenes ni mensajes largos directamente. Por eso es qu esolo proteje la clave AES que es la que cigra los datos reales de cualquier tamaño.

3. Cifrado y Descifrado Directo con RSA-OAEP
    c. ¿Porqué cifrar el mismo mensaje dos veces produce resultados distintos? Demuéstrenlo y expliquen que propiedad de OAEP lo cause
    Genera un cifrado distinto con el mismo mensaje dado que en OAEP existe aleatoriedad, por lo cual en cada ejecucion de cifrado se genera un seed aleatorio que se aplica al Data Block

    DB = Hash(label) || padding || mensaje

    en el cual con la seed aleatoria se usa para generar una mascara que se usa para crear un bloque final en el cual es el bloque final que si se usa para cifrar.

    EM = 0x00 || maskedSeed || maskedDB

    En el cual a este se aplica el cifrado 

    ciphertext = EM^e mod n
